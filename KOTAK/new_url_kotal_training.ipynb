```python
import requests
from bs4 import BeautifulSoup
# import uuid
import os
```

```python
# Function to scrape content from a URL and save it to a text file
def scrape_and_save(url, uuid):
    # Fetch content from the URL
    response = requests.get(url)
    if response.status_code == 200:
        # Parse HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        content = soup.get_text()
        
        # Generate a UUID for the content
#         file_uuid = str(uuid.uuid4())
        file_uuid = uuid
        
        # Save content to a text file with UUID as filename
        with open(f"{file_uuid}.txt", "w", encoding='utf-8') as f:
            f.write(content)
        
        return file_uuid
    else:
        print(f"Failed to fetch data from {url}")
        return None

# Example URLs and their corresponding UUIDs (you can replace these with your own data)
uuid_url_mapping = {
    "2160dc86-070f-4728-8a89-23adff34596e": "https://www.sebi.gov.in/sebiweb/home/HomeAction.do?doListing=yes&sid=1&ssid=4&smid=0",
    "badb0567-20d8-44a1-a2a8-7946c36a5a11": "https://www.sebi.gov.in/sebiweb/home/HomeAction.do?doListing=yes&sid=2&ssid=11&smid=0",
    "9731a569-604f-45cb-9359-30c8c0f37818": "https://www.sebi.gov.in/sebiweb/other/OtherAction.do?doMutualFund=yes&mftype=2",
    "462ed81c-4e2b-4f4b-82bc-f96db6cda7b9": "https://www.sebi.gov.in/sebiweb/home/cause-list.jsp",
    "2f16cc40-6ee5-4398-819e-399edf6382dc": "https://www.sebi.gov.in/sebiweb/home/HomeAction.do?doListing=yes&sid=6&ssid=25&smid=0"
}

# Scrape content from each URL and store its UUID
uuid_content_mapping = {}
for uuid, url in uuid_url_mapping.items():
    content_uuid = scrape_and_save(url, uuid)
    if content_uuid:
        uuid_content_mapping[uuid] = (content_uuid, url)

# Function to retrieve content and its URL based on user queries
def get_content_by_query(query):
    for uuid, (content_uuid, url) in uuid_content_mapping.items():
        with open(f"{content_uuid}.txt", "r") as f:
            content = f.read()
            if query in content:
                return content, url
    return "No matching content found", None
```
